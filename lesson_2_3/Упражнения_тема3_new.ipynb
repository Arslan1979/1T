```python
# перегоняем их в pd
data = pd.DataFrame().from_dict(data_dict)
data
```

```python
# обучаем модель
features = data.drop(['city_name', 'is_wch_city'], axis=1)
target = data['is_wch_city']
model = DecisionTreeClassifier()
model.fit(features, target)
```

```python
features.columns
```

```python
# выводим дерево
plot_tree(model, feature_names=list(features.columns));
```

```python
# рассчитываем важность признаков
pd.DataFrame(
    zip(
        features.columns, 
        model.feature_importances_
    ), 
    columns=['names', 'importances']
).sort_values(by='importances', ascending=False)
```

```python
def my_factorial(num):
    
    if num == 1:
        return 1
    
    prev_factorial = my_factorial(num-1)
    print(num, prev_factorial)
    return num * prev_factorial
```
